##### Vector

* We write ${\bf v}=(a,b,c)$ and call the numbers $a, b$ and $c$ the components of the vector $\bf v$.

##### $\mathbb{R}^3$ as a Vector Space
Let $\bf u, v$ and $\bf w$ be arbitrary vectors in $\mathbb{R}^3$, and $a$ and $b$ scalars in $R$, then
1. $\bf u + v = v + u$ (Commutativity of addition)
2. $\bf u + (v + w) = (u + v) + w$ (Associativity of addition)
3. There exists an element $0 \in \mathbb{R}^3$, called the zero vector, such that $v + 0 = v$ for all $v \in \mathbb{R}^3$.
4. For every ${\bf v} \in \mathbb{R}^3$, there exists an element ${\bf −v}\in \mathbb{R}^3$, called the additive inverse of $\bf v$, such that $\bf v + (−v) = 0$.
5. $a({\bf u} + {\bf v}) = a{\bf u} + a{\bf v}$.
6. $(a + b){\bf v} = a{\bf v} + b{\bf v}$.
7. $a(b{\bf v}) = (ab){\bf v}$.
8. $1({\bf v}) = {\bf v}$, where 1 denotes the multiplicative identity in $\mathbb{R}^3$.

#### Linear Independence in $\mathbb{R}^3$.
The three vectors $\bf u, v$, and $\bf w \in \mathbb{R}^3$ are said to __linearly dependent__ provided that one of them is a linear combination of the other two; that, either 
$$
\begin{array}{rrl}
u &=& rv + sw \\
v &=& ru + sw \\
w &=& ru + sv 
\end{array}
$$
for some $r, s \in \mathbb{R}$. The vectors are said to be __linearly independent__ if the are not linearly indepedent.

#### Theorem
The three vectors $\bf u, v$ and $w$ in $\mathbb{R}^3$ are linearly dependent if and only if there exists scalars $a, b$ and $c$ *NOT ALL ZEROS* such that 
$$
a{\bf u} + b{\bf v} + c{\bf w} ={\bf 0}.
$$

* The three vectors $\bf u, v$ and $w$ in $\mathbb{R}^3$ are linearly independent if and only if 
$$
a{\bf u} + b{\bf v} + c{\bf w} ={\bf 0}.
$$
implies that $a=b=c=0$.
* The three vectors ${\bf u}=(u_1,u_2,u_3), {\bf v}=(v_1,v_2,v_3)$ and ${\bf w}=(w_1,w_2,w_3)$ in $\mathbb{R}^3$ are linearly independent if and only if 
$$
\left|
\begin{array}{ccc}
u_1&u_2&u_3\\
v_1&v_2&v_3\\
w_1&w_2&w_3
\end{array}
\right| \not = 0
$$

#### Basis for $\mathbb{R}^3$
A __basis__ for $\mathbb{R}^3$ is a triple $\bf u, v, w$ of vectors such that evey $t\in \mathbb{R}^3$ can be expressed as a linear combination
$$
t = a{\bf u} + b{\bf v}+ c{\bf w}
$$
where $a,b,c \in \mathbf{R}$.

#### Theorem
If the vectors $\bf u, v$ and $w$ in $\mathbb{R}^3$ are linearly independent, then they constitute a basis for $\mathbb{R}^3$.
